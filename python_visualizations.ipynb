```python
mycursor = connection.cursor()
```

```python
mycursor.execute("select distinct p.Fname as Students,t.grade as Grades from person p, transcript t where t.student_id=p.SSN ORDER BY Students LIMIT 10")
result = mycursor.fetchall
```

```python
Students = []
Grades = []

for i in mycursor:
    Students.append(i[0])
    Grades.append(i[1])
   
print("Students=",Students)
print("Grades = ",Grades)
```

```python
plt.bar(Students, Grades,color=('r','g'))

plt.xlabel('Name of Students')
plt.ylabel('Grades of Students')
plt.title('Performance of Students')
plt.rcParams["figure.figsize"] = 10,4.5
plt.figure(figsize=(10,4.5))
```

```python
mycursor.execute("select distinct count(student_id) as Number_of_Students,d.dname as Department from student s, department d where s.major=d.dname group by d.dname")
result = mycursor.fetchall
```

```python
Number_of_Students = []
Department = []

for i in mycursor:
    Number_of_Students.append(i[0])
    Department.append(i[1])
   
print("Students=",Number_of_Students)
print("Grades = ",Department)
```

```python
plt.pie(Number_of_Students,labels=Department,autopct='%1.1f%%', pctdistance=0.85)

plt.show()
```
